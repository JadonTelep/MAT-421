### Section 3.1.1 (Limits and Continuity)

Limits are essential to calculus and mathematical analysis, and are used to
define continuity, derivatives, and integrals.

In this chapter, we will use the **Euclidean norm** $\|x\| = \sqrt{\sum_{i=1}^dx_i^2}$ for $x=(x_1,...,x_d)^T \in ℝ^d$.

The open $r$-ball around $x \in ℝ^d$ is the set of points within Euclidean discance $r$ of $x$:

$B_r(x) = \{y \in ℝ^d : \|y-x\|<r\}$

A point $x \in ℝ^d$ is a **limit point** (or accumulation point) of a set $A \subseteq ℝ^d$ if every open ball around $x$ contains an element $a$ of $A$ such that $a \neq x$ as in **Figure 3.1**.

A set $A$ is **closed** if every limit point of $A$ belongs to $A$.

A set $A$ is **open** if there is a $B_r(x) \subseteq A$ for all $ x \in A$ as in **Figure 3.2**.

A set $A \subseteq ℝ^d $ is **bounded** if there exists an $r>0$ such that $A \subseteq B_r(0)$, where $0=(0,...,0)^T$

**Limits of a function** - Let $f:D → ℝ$ be a real-valued function on $D \in ℝ^d$. Then $f$ is said to have a limit $L \in ℝ$ as $x$ approaches $a$ if: for any $𝛆 > 0$, there exists a $δ>0$ such that $|f(x) - L| < ϵ$ for all $x \in D ∩B_δ(a)$

$\lim_{x \to a}f(x)=L$

Continuous function is a function that does not have any abrupt changes in value, known as discontinuities. The definition of continuity of a function can be visualized in **Figure 3.3**. Note that we explicitly exclude $a$ itself from having to satisfy the condition $|f(x) - L| < ε$. In particular, we may have $f(a) ̸ = L$. We also do not restrict $a$ to be in $D$.

**Continuous function** - Let $f:D → ℝ$ be a real-valued function on $D \in ℝ^d$. Then $f$ is said to **continuous** at $ a \in D$ if

$\lim_{x \to a}f(x)=f(a)$

### Section 3.1.2 (Derivatives)

**Derivative** - Let $f:D → ℝ$ where $D ⊆ ℝ$ and let $x_0 \in D$ be an interior point of $D$. The **derivative** of $f$ at $x_0$ is

$f'(x_0) = \frac{df(x_0)}{dx} = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}$

provided the limit exists. See **Figure 3.5**

Many functions in data science involves several independent variables. For
functions of several variables, we have the following generalization. As before,

**Partial Derivative** - Let $f : D \to ℝ$ where $D ⊆ ℝ^d$ and let $x_0 \in D$ be an interior point of $D$. The **partial derivative** of $f$ at $x_0$ with respect to $x_i$ is

$\frac{∂f(x_0)}{∂x} = \lim_{h \to 0} \frac{f(x_0 + he_i) - f(x_0)}{h}$

provided the limit exists. If $\frac{∂f(x_0)}{∂x}$ exists and is continuous in an open ball around $x_0$ for all $i$, then we say that $f$ continuously differentiable at x_0.

## Section 3.2 (Taylor's Theorem)

Taylor's theorem gives an approximation of a differentiable function around
a given point by a polynomial. We will make use of Taylor's Theorem, a powerful generalization of the Mean Value Theorem that provides polynomial approximations to a function around a point.

**Taylor** -  Let $f : D \to ℝ$ where $D ⊆ ℝ$ Suppose $f$ has a $m$ times continuous derivative on $[a,b]$. Then

$f(b) = f(a) + (b-a)f'(a) + \frac{1}{2}(b-a)^2f''(a)+...+\frac{(b-a)^{m-1}}{(m-1)!}f^{(m-1)}(a) + R_m$

where $R_m = \frac{(b-a)^{m}}{(m)!}f^{m}(a + θ(b-a))$ for some $0<θ<1$

## Section 3.3 (Gradient Decent)

**Gradient descent** - Gradient descent is an iterative optimization algorithm for finding a local minimum of a differentiable function. Once we know a function has a minimizer, we will discuss a class of algorithms known as gradient descent method for solving optimization problems numerically. Let $f : ℝ^d → ℝ$ is continuously differentiable. We restrict ourselves to unconstrained minimization problems of the form

$\min_{x\in ℝ^d}f(x)$

One might evaluate $f$ at a large number of points $x$ to identify a global
minimizer of $f$. This naive approach seems too expensive. A less naive ap-
proach might be to find all stationary points of $f$, that is, those $x$'s such that $∇ f(x) = 0$. And then choose that $x$ among them that produces the smallest value of $f(x)$.

**Steepest Descent** - Let $f : ℝ^d → ℝ$ be continuously differentiable at $x_0$. For any unit vector $v \in ℝ^d$,

$\frac{∂f(x_0)}{∂v} \ge \frac{∂f(x_0)}{∂v^*}$ where $v^* = -\frac{∇f(x_0)}{∇f\|(x_0)\|}$

At each iteration of steepest descent, we take a step in the direction of the
negative of the gradient, that is,

$x^{k+1} = x^k - \alpha_k∇f(x^k), k=0,1,2...$

for a sequence of steplengths $αk > 0$. $αk$ are called step sizes. In general, we will not be able to guarantee that a global minimizer is reached in the limit, even if one exists.
